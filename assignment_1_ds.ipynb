```python
import pandas as pd
data=pd.read_csv("/content/sales_data_with_discounts.csv")
num=data.select_dtypes(include='number')
print("mean\n",num.mean())
print("median\n",num.median())
print("mode\n",num.mode().iloc[0])
print("standard deviation\n",num.std())
```

 Interpretation of the Statistics:
1. Volume
Mean: On average, about 5.07 units are sold per transaction.

Median: Half of the transactions involve selling 4 units or less.

Mode: The most common volume sold is 3 units.

Std Dev: A value of 4.23 shows there's some variation, but most transactions involve a small number of units.

2. Average Price
Mean: The average price across items is ₹10,453.43, which is high.

Median: The middle price is much lower, ₹1,450, suggesting some very expensive items (outliers) increased the mean.

Mode: The most common price is ₹400, meaning most products are low-priced.

Std Dev: A very high value (₹18,079.90) indicates large variation in item prices.

3. Total Sales Value
Mean: The average total value of a sale is ₹33,812.84.

Median: Half the sales are less than ₹5,700, again showing the influence of big sales pulling the mean up.

Mode: The most common sales value is ₹24,300.

Std Dev: ₹50,535.07 shows that sales values vary a lot.

4. Discount Rate (%)
Mean: On average, a 15.16% discount is applied.

Median: The middle discount is 16.58%, close to the mean — indicating consistency.

Mode: The most common discount is 5.01%, possibly a standard low offer.

Std Dev: 4.22% shows moderate variation in discount rates.

5. Discount Amount
Mean: On average, customers receive ₹3,346.50 as discount.

Median: Most discounts are less than ₹988.93 — again, outliers are increasing the mean.

Mode: Most frequently, customers get around ₹69.18 off.

Std Dev: A high value (₹4,509.90) means big variation in discount amounts.

6. Net Sales Value
Mean: After discounts, the average sale is worth ₹30,466.34.

Median: Half of the sales are under ₹4,677.79, showing skewed data.

Mode: The most common net value is ₹326.97, showing many small-value sales.

Std Dev: ₹46,358.66 means sales amounts vary widely, likely due to high-priced items.
